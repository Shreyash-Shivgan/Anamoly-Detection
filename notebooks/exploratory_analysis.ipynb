# Exploratory Data Analysis

This notebook provides a basic analysis of the power consumption dataset to visualize seasonality, trends, and obvious consumption spikes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style="whitegrid")
```

```python
# Load data (run src/train.py first to generate synthetic data if you don't have a dataset)
data_path = "../data/raw/data.csv"
if os.path.exists(data_path):
    df = pd.read_csv(data_path)
    print(df.head())
else:
    print(f"File '{data_path}' not found.")
```

```python
if 'df' in locals():
    # Attempt to plot the first numeric column
    numeric_cols = df.select_dtypes(include='number').columns
    if len(numeric_cols) > 0:
        col = numeric_cols[0]
        plt.figure(figsize=(15, 6))
        plt.plot(df.index, df[col], alpha=0.7)
        plt.title(f"Raw Time Series - {col}")
        plt.xlabel("Index")
        plt.ylabel(col)
        plt.show()
```
